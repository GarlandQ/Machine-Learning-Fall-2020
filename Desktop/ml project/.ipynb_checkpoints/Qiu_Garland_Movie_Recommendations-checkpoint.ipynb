```python
# TMDB 5000 dataset
tmdb5000_df1 = pd.read_csv('tmdb_5000_movies.csv')
print(tmdb5000_df1.shape)
tmdb5000_df1.head(5)
```

```python
# Checking if we're missing any null values intially
tmdb5000_df1.isna().sum()
```

```python
# Based on the output, homepage, overview, release_date, runtime, and tagline have null values here. 
# Homepage, release_date, and runtime will most likely not have any significant result in training and testing.
# We do want to fill in empty strings for the null values in our dataset.

tmdb5000_df1['homepage'] = tmdb5000_df1['homepage'].fillna('')
tmdb5000_df1['overview'] = tmdb5000_df1['overview'].fillna('')
tmdb5000_df1['release_date'] = tmdb5000_df1['release_date'].fillna('')
tmdb5000_df1['tagline'] = tmdb5000_df1['tagline'].fillna('')

# For the null runtimes, I will be using the runtimes' mean to fill.
tmdb5000_df1['runtime'] = tmdb5000_df1['runtime'].fillna(tmdb5000_df1['runtime'].mean())
```

```python
# Checking if we're missing any null values afterwards
tmdb5000_df1.isna().sum()
```

### Cleaning the data more to run models

```python
# One of the problems in the dataset are the JSON formatted features.
# We need to change these into strings that can be used for our models.

# features = ['cast', 'crew', 'genres', 'keywords', 'production_companies']
# Cleaning cast column
df['cast'] = df['cast'].apply(json.loads)
for index, j in zip(df.index, df['cast']):
    lst = []
    for k in range(len(j)):
        lst.append((j[k]['name']))
    df.loc[index, 'cast'] = str(lst)


# Cleaning crew column
# Only going to take director
df['crew'] = df['crew'].apply(json.loads)
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
df['crew'] = df['crew'].apply(get_director)


# Cleaning genres
df['genres'] = df['genres'].apply(json.loads)
for index, j in zip(df.index, df['genres']):
    lst = []
    for k in range(len(j)):
        lst.append((j[k]['name']))
    df.loc[index, 'genres'] = str(lst)
    

# Cleaning keywords
df['keywords'] = df['keywords'].apply(json.loads)
for index, j in zip(df.index, df['keywords']):
    lst = []
    for k in range(len(j)):
        lst.append((j[k]['name']))
    df.loc[index, 'keywords'] = str(lst)


# Cleaning production_companies
df['production_companies'] = df['production_companies'].apply(json.loads)
for index, j in zip(df.index, df['production_companies']):
    lst = []
    for k in range(len(j)):
        lst.append((j[k]['name']))
    df.loc[index, 'production_companies'] = str(lst)
```

### Content Based Filtering
In this recommendation system, the content of the movies, such as overview, cast, crew, keywords, and taglines, are used to find the similarity with other movies.

### Extracting overviews using linear kernel to recommend movies

```python
# Looking at a random overview
df['overview'][np.random.choice(df.shape[0])]
```

```python
# Keeping "important" words and removing common words.
tfidf = TfidfVectorizer(stop_words='english',analyzer='word')
tfidf_matrix = tfidf.fit_transform(df['overview'])
print(tfidf_matrix.shape)
tfidf_matrix
```

```python
similarity = linear_kernel(tfidf_matrix, tfidf_matrix)
print(similarity.shape)
similarity
```

```python
indices = pd.Series(df.index, index=df['original_title'])

def overview_recommendation(title, sim=similarity):
    idx = indices[title]
    similar = list(enumerate(sim[idx]))
    similar = sorted(similar, key=lambda x:x[1], reverse=True)
    similar = similar[1:11]
    m_idx = [i[0] for i in similar]
    return m_idx, idx


random_movie = df['original_title'][np.random.choice(df.shape[0])]
movie_index,idx=overview_recommendation(random_movie)
print("Movie Picked:", random_movie)
for i in movie_index:
    print(df['original_title'].iloc[i],'(content match rate):',similarity[idx][i])

```

Based on these results, recommending movies based on context would be very weak, as we are only basing this off overviews, which typically is a hookline for people to watch a movie. However, this does not mean it is not unreliable, as some results do help recommend other movies similar to it.

### Using KNN with genre and the cast to recommend movies

```python
# Cleaning genre list
df['genres'] = df['genres'].str.strip('[]').str.replace(' ','').str.replace("'",'')
df['genres'] = df['genres'].str.split(',')

# Cleaning cast list
df['cast'] = df['cast'].str.strip('[]').str.replace(' ','').str.replace("'",'').str.replace('"','')
df['cast'] = df['cast'].str.split(',')
```

```python
# Checking the unique genres in the list
genre_list = []
for index, row in df.iterrows():
    genres = row["genres"]
    
    for genre in genres:
        if genre not in genre_list:
            genre_list.append(genre)
    
genre_list
```

```python
# Cleaning the cast data up a little more
# Another thing is to take a few of the actors, mainly the star roles of each movie, rather than all of the actors.
# In this case, I will take the 10 main cast.
for i,j in zip(df['cast'],df.index):
    lst = []
    lst = i[:10]
    df.loc[j,'cast'] = str(lst)
df['cast'] = df['cast'].str.strip('[]').str.replace(' ','').str.replace("'",'')
df['cast'] = df['cast'].str.split(',')
for i,j in zip(df['cast'],df.index):
    lst = []
    lst = i
    lst.sort()
    df.loc[j,'cast'] = str(lst)
df['cast'] = df['cast'].str.strip('[]').str.replace(' ','').str.replace("'",'')
```

```python
df['cast']
```

```python
# Checking the unique cast members in a list
# cast_list = df.cast.unique().tolist() 
# cast_list
cast_list = []
for index, row in df.iterrows():
    cast = row["cast"]
    
    for i in cast:
        if i not in cast_list:
            cast_list.append(i)
```

```python
# Forming a binary list of the genres that the movie contains to later use for our KNN model.
def binary_genre_lst(gl):
    lst = []
    
    for genre in genre_list:
        if genre in gl:
            lst.append(1)
        else:
            lst.append(0)
    
    return lst

# We can add this to our dataset
df['binary_genres'] = df['genres'].apply(lambda x: binary_genre_lst(x))
df['binary_genres']
```

```python
# Forming a binary list of the cast that the movie contains to later use for our KNN model.
def binary_cast_lst(cl):
    lst = []
    for c in cast_list:
        if c in cl:
            lst.append(1)
        else:
            lst.append(0)
    
    return lst

# We can add this to our dataset
df['binary_cast'] = df['cast'].apply(lambda x: binary_cast_lst(x))
df['binary_cast']
```

```python
df.head(5)
```

```python
# Calculate the distance between two points using cosine similarity
# The more the distance, the less similarity they have
def genre_cast_cosine_similarity(id1, id2):
    a = df.iloc[id1]
    b = df.iloc[id2]
    
    genres_a = a['binary_genres']
    genres_b = b['binary_genres']
    
    genresDistance = spatial.distance.cosine(genres_a, genres_b)
    
    cast_a = a['binary_cast']
    cast_b = b['binary_cast']
    castDistance = spatial.distance.cosine(cast_a, cast_b)
    
    return genresDistance + castDistance


# Looping to see more movies on output
for n in range(9):
    # Checking the similarities between two random movies
    random1 = np.random.choice(df.shape[0])
    random2 = np.random.choice(df.shape[0])
    print(genre_cast_cosine_similarity(random1, random2))

    # Checking what those random movies were
    print(df.iloc[random1].original_title)
    print(df.iloc[random2].original_title)
    
    print()
```

```python
# Recommend function
def recommend(name):
    # Match movie with a name similar to it from the dataset
    match_movie = df[df['original_title'].str.contains(name)].iloc[0].to_frame().T
    print('Selected Movie: {} \n'.format(match_movie.original_title.values[0]))
    def getNeighbors(original_movie, k):
        distances = []
    
        for index, movie in df.iterrows():
            if movie['index'] != original_movie['index'].values[0]:
                sim = genre_cast_cosine_similarity(original_movie['index'].values[0], movie['index'])
                distances.append((movie['index'], sim))
    
        distances.sort(key=operator.itemgetter(1))
        # You can see the distances of all the movies comparing with the parameter below.
        #print("Distances:",distances)
        neighbors = []
    
        for x in range(k):
            neighbors.append(distances[x])
            print("Distance:",distances[x])
        return neighbors

    k = 10
    neighbors = getNeighbors(match_movie, k)
    
    print('\nRecommended Movies:\n')
    for neighbor in neighbors:
        print("Movie Name: {}\n Average Rating: {} \t Genre: {}".format(df.iloc[neighbor[0]][7],
                                                                           df.iloc[neighbor[0]][18],
                                                                           df.iloc[neighbor[0]][2]))
        print('\n')

    print('------------------------------------------------------------------------------------------------------------\n')


# Test recommend by providinig a random movie
# Looping to see more recommendations
for i in range(2):
    pick_movie = np.random.choice(df.original_title.sample())
    recommend(pick_movie)
```

```python
# Putting into list and then converting to binary based on list
keyword_list = []
for index, row in df.iterrows():
    keyword = row["keywords"]
    
    for k in keyword:
        if k not in keyword_list:
            keyword_list.append(k)


def binary_keywords_lst(kw):
    lst = []
    for k in keyword_list:
        if k in kw:
            lst.append(1)
        else:
            lst.append(0)
    return lst


# We can add this to our dataset
df['binary_keywords'] = df['keywords'].apply(lambda x: binary_keywords_lst(x))
df['binary_keywords']
```

### Using KNN with genres, cast, director, and keyword to recommend movies
